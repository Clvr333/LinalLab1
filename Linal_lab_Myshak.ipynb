```python
def is_palindrome(n: int) -> bool:

  reversed_n = int(str(n)[::-1])

  return reversed_n == n


def get_primes(end: int) -> List[int]:

  nums = [True for num in range(end + 1)]

  for div in range(2, int((end + 1)**0.5) + 1):
    if nums[div]:
      for num in range(div**2, end + 1, div):
        nums[num] = False

  primes = [num for num in range(end + 1) if nums[num]]

  return primes[2:]


def get_cyclic_permutations(n: int) -> List[int]:

  perms = []
  str_n = str(n)

  for shift in range(len(str_n)):
    str_n = str_n[-1] + str_n[:-1]
    perms.append(int(str_n))

  return perms


def is_prime(n: int) -> bool:

  if n < 2:
    return False

  for div in range(2, int(n**0.5) + 1):
    if n % div == 0:
      return False

  return True

def get_numbers_with_two_digits(digits: str) -> List[int]:

  digits += '0'
  result = []

  for digit_one in digits:
    for digit_two in digits:
      for digit_three in digits:
        for digit_four in digits:
          for digit_five in digits:
            for digit_six in digits:
              for digit_seven in digits:
                for digit_eight in digits:
                  for digit_nine in digits:
                    for digit_ten in digits:

                      num = digit_one + digit_two + digit_three + digit_four + digit_five + \
                           digit_six + digit_seven + digit_eight + digit_nine + digit_ten
                      flag = 0

                      for digit in digits[:-1]:
                        if digit + '0' in num:
                          flag = 1
                          break

                      if flag == 0:
                        result.append(int(num))

  result.sort()

  return result[1:]


def pi(n: int) -> int:

  return len(get_primes(n))


def pi_2(n: int) -> int:

  count_pairs = 0
  primes = get_primes(n)

  for index in range(len(primes) - 1):
    if primes[index + 1] - primes[index] == 2:
      count_pairs += 1

  return count_pairs


def factorial(n: int) -> int:

  result = 1

  if n != 1:
    result *= (n * (factorial(n-1)))

  return result
```

```python
def palindromic_squares_and_circular_primes() -> tuple[List[int], List[int]]:
  """
  Возвращает:
  tuple:
  - список всех палиндромов a < 100000, для которых a^2 — палиндром;
  - список всех простых p < 1000000, все циклические перестановки цифр которых просты.
  """
  palindromic_squares = []

  for a in range(1, 10**5):
    if is_palindrome(a) and is_palindrome(a**2):
      palindromic_squares.append(a)

  circular_primes = []
  primes = get_primes(10**6)

  for num in primes:
    perms = get_cyclic_permutations(num)
    flag = 0
    for perm in perms:
      if not is_prime(perm):
        flag = 1
        break
    if flag == 0:
      circular_primes.append(num)

  return (palindromic_squares, circular_primes)
```

```python
def palindromic_cubes_and_palindromic_primes() -> tuple[List[int], List[int]]:
  """
  Возвращает:
  tuple:
  - список всех палиндромов a < 100000, для которых a^3 — палиндром;
  - список всех простых p <= 10000, которые являются палиндромами.
  """
  palindromic_cubes = []
  for a in range(1, 10**5):
    if is_palindrome(a) and is_palindrome(a**3):
      palindromic_cubes.append(a)

  primes = get_primes(10**4)
  palindromic_primes = []

  for prime in primes:
    if is_palindrome(prime):
      palindromic_primes.append(prime)

  return (palindromic_cubes, palindromic_primes)
```

```python
def primes_with_two_digits() -> Dict[str, List[int]]:
  """
  Возвращает словарь вида:
  {
  '13': [список первых 100 простых из {1,3}],
  '15': [список первых 100 простых из {1,5}],
  '17': [список первых 100 простых из {1,7}],
  '19': [список первых 100 простых из {1,9}]
  }
  """

  result = {'13': [],
            '15': [],
            '17': [],
            '19': []}

  keys = ['13', '15', '17', '19']
  for key in keys:
    two_digits_nums = get_numbers_with_two_digits(key)

    for num in two_digits_nums:
      if is_prime(num):
        result[key].append(num)
      if len(result[key]) == 100:
        break

  return result
```

```python
def twin_primes_analysis(limit_pairs: int = 1000) -> Tuple[List[Tuple[int, int]], List[float]]:
  """
  Возвращает:
  - список первых `limit_pairs` пар близнецов (p, p+2);
  - список значений отношения pi_2(n) / pi(n) для n, соответствующих последним
  ,→ элементам каждой пары,
  где pi_2(n) — количество пар близнецов <= n, pi(n) — количество простых <= n.
  """
  primes = get_primes(10**6)
  pairs = []
  ratios = []

  for index in range(len(primes)):
    if primes[index + 1] - primes[index] == 2:
      pairs.append((primes[index], primes[index + 1]))
      ratios.append(pi_2(primes[index + 1]) / pi(primes[index + 1]))
    if len(pairs) == limit_pairs:
      break

  return (pairs, ratios)
```

```python
def factorial_plus_one_factors() -> Dict[int, Dict[int, int]]:
  """
  Возвращает словарь вида:
  { n: {простой_делитель: степень, ...}, ... }
  для n от 2 до 50, где ключ — n, значение — разложение n! + 1 на простые множители.
  """
  result = {}

  for n in range(2, 51):
    fact = factorial(n) + 1
    ota = factorint(fact)
    result[fact] = ota

  return result
```

```python
def gcd(a, b):

  while b != 0:
    a, b = b, a % b

  return a
```

```python
def euler_phi_direct(n: int) -> int:
  """Вычисляет (n) прямым перебором."""

  counter = 0

  for k in range(1, n + 1):
    if gcd(n, k) == 1:
      counter += 1

  return counter


def euler_phi_factor(n: int) -> int:
  """Вычисляет (n) через разложение на простые множители."""

  ota = factorint(n)
  result = n
  factors = ota.keys()

  for factor in factors:
    result *= (1 - 1 / factor)

  return result


def compare_euler_phi_methods(test_values: List[int]) -> dict:
  """
  Сравнивает время работы трёх методов на заданных значениях.
  Возвращает словарь с тремя списками времён (в секундах).
  """
  result = {'direct': [], 'factor': [], 'totient': []}
  for n in test_values:

    start_time = time.time()
    euler_phi_direct(n)
    end_time = time.time()
    result['direct'].append(end_time - start_time)

    start_time = time.time()
    euler_phi_factor(n)
    end_time = time.time()
    result['factor'].append(end_time - start_time)

    start_time = time.time()
    totient(n)
    end_time = time.time()
    result['totient'].append(end_time - start_time)

  return result
```

```python
def subgroups_of_Sm(N: int) -> dict:
  S = SymmetricGroup(m)
  subs = S.subgroups()
  count = len(subs)
  print(f"Количество подгрупп: {count}")

  rnd_sub = rd.choice(subs)
  print(f"Случайная подгруппа (порядок {rnd_sub.order()}): {rnd_sub.gens()}")

  idx = N % count
  target_H = subs[idx]
  print(f"Выбрана подгруппа с индексом {idx} (из списка): {target_H.gens()}")

  left_cosets = S.cosets(target_H, side='left')
  right_cosets = S.cosets(target_H, side='right')

  print(f"Левые смежные классы (всего {len(left_cosets)}): {left_cosets[:3]} ...")
  print(f"Правые смежные классы (всего {len(right_cosets)}): {right_cosets[:3]} ...")

  h_index = S.order() / target_H.order()
  is_norm = target_H.is_normal(S)

  print(f"Индекс подгруппы: {h_index}")
  print(f"Является нормальной: {is_norm}")

  return {"count": count, "target_subgroup": target_H, "is_normal": is_norm}
```

```python
def solve_sigma_power_eq(N: int) -> dict:
  S = SymmetricGroup(m)
  target_tuple = (1, 2, 3, 4)
  target_cycle = S(target_tuple)
  print(f"Целевой цикл: {target_cycle}")

  solutions = []
  for sigma in S:
    if sigma^n == target_cycle:
      solutions.append(sigma)

  print(f"Количество решений: {len(solutions)}")

  if solutions:
    show_count = min(3, len(solutions))
    random_sol = rd.sample(solutions, show_count)
    print(f"Случайные 3 решения: {random_sol}")
  else:
    print("Решений нет")

  return solutions
```

```python
def subgroups_of_Zm_star(N: int) -> list:
  units = [x for x in IntegerModRing(m) if x.is_unit()]
  print(f"Элементы Z*_{m}: {units}")

  valid_subgroups = []

  identity = IntegerModRing(m)(1)

  for r in range(1, len(units) + 1):
    for subset in itertools.combinations(units, r):
      if identity not in subset:
        continue

      is_closed = True
      for a in subset:
        for b in subset:
          if (a * b) not in subset:
            is_closed = False
            break
        if not is_closed:
          break

      if is_closed:
        valid_subgroups.append(subset)

  print(f"Найденные подгруппы: {valid_subgroups}")
  return valid_subgroups
```

Проверка (f * h) mod g: 1

```python
def generate_irreducible_polynomials(q: int, d: int) -> list:
  """
  Возвращает список всех неприводимых полиномов степени d над F_q.
  """
  F = GF(q)
  R = PolynomialRing(F, 'x')
  x = R.gen()

  irreducibles = []

  for coeffs in itertools.product(range(q), repeat=d):
    poly = x^d
    for power, coeff_val in enumerate(coeffs)
      poly += F(coeff_val) * (x**power)

    if poly.is_irreducible():
      irreducibles.append(poly)

  return irreducibles
```

```python
test_cases = [(2, 2), (3, 3), (5, 4)]

for q_val, d_val in test_cases:
  result = generate_irreducible_polynomials(q_val, d_val)
  print(f"F_{q_val}, степень {d_val}: {len(result)} неприводимых.")
  print(f"{result[:5]} ...\n")
```
